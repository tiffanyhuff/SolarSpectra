```python
import csv
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt(r'C:\Users\tiff\Documents\Modern1\1_4\spectrum0.txt', delimiter='\t', unpack=True)
plt.plot(x,y, label='Spectral Data')

plt.xlabel('Wavelength [nm]')
plt.ylabel('Intensity [counts]')
plt.title('Solar Spectrum at 1.4 Degree Altitude')
plt.legend()
plt.show()
xmax = x[np.argmax(y)]
ymax = y.max()

print(xmax,ymax)
```
